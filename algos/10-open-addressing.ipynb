# 10. Open Addressing, Cryptographic Hashing

- Open addressing
    - another way to implement a hash table other than hashing with chaining
    - no chianing, no pointers
    - way to implement hash tables using arrays
    - way it works:
        - 1 item per slot
        - m >= n
            - m = slots
            - n = elements
    - IDEA: Probing
        - we need a hash function that specifies order of slots to probe (try) for  a key (for insert/ search/ delete)
        - h: U x {0, 1, ..., m - 1} --> {0, 1, ..., m - 1}
            - U = universe of keys
            - {0, 1, ..., m - 1} = trial count (as argument)
                - you will go through different trial to get a valid position
            - {0, 1, ..., m - 1} = slot number (as result)
    - We want all slot to be able to get filled out
        - want h(k, 1), h(k, 2), ... h(k, m-1) to be permutations of 0, 1, ... m-1
    - Methods:
        - insert(k, v)
            - keep probing until an empty slot is found. insert item when empty is found.
                - insert treats DeleteMe the same as None
        - search(k)
            - as long as the slots encountered are occupied by keys != k, keep probing until you either encounter k or find an empty slot
                - sucess means finding k 
                - encountering a DeleteMe flag, search continues
                - failure means finding empty or not finding it at all
        - delete(k)
            - replaced delete item with DeleteMe flag (a different flag than None
    - Probing strategies/ modifying hash functions:
        - Linear probing: h(k, i) = (h'(k) + i) mod m
            - clarification:
                - h' is a regular hash function
                - mod m satisfies the permutation
            - PROBLEM: clusters
                - clustering is consecutive groups of occupied slots which keep growing (they get longer and longer)
        - Double hashing: h(k, i) = (h1(k) + i * h2(k)) mod m
            - if h2(k) is relatively prime to m --> we get the permutations we want
                - m = 2^r, h2(k) for all k is odd
    - Uniform Hashing Assumption (!= simple uniform hashing)
        - each key is equally likely to have any one of the m! permutations as its probe sequence
            - with this we can prove that the cost of operations for insert, search, delete is <= 1/(1-alpha)
                - alpha = n/m
