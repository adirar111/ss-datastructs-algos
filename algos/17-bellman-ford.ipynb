# 17. Bellman-Ford

- Bellman-Ford shortest path algorithm works for graphs with negative cycles.
    - it reports the existence of a negative cycle and avoids this computation
- Recall Generic Shortest Path Algorithm:
    - Problems:
        - Complexity could be exponential
            - SOLUTION: Dijkstra
        - Algorithm will not terminate if there is a negative weight cycle
            - SOLUTION: Bellman-Ford
- Bellman-Ford Pseudocode:
    - ``` 
        Bellman-Ford(G, W, s)
            Initialize() --> sets d values for vertices
            for i = 1 to |V| - 1:
                for each edge (u,v) in E
                    Relax(u, v, w)
            # below is a check for a negative weight cycles
            for each edge(u, v) in E
                if d[v] > d[u] + w(u, v)
                    report that negative cycle exists
        ```
    - Complexity is O(VE + E) which is O(VE) --> polynomial time
- Analysis of BF:
    - Theorem: IF G(V, E) contains no negative weight cycles then after BF executes, d[v] = S(s, v) for all v in V
    - Corollary: If a value d[v] fails to converge after V-1 passes (first for block), there exists a negative weight cycle reachable from s.
