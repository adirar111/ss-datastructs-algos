# 9. Table Doubling, Karp-Rabin

- How do we choose m?
    - When making a table, we need to make a table of size m. How do we know what to make m?
    - We want m = O(n) or proportional to n (theta)
        - we know alpha = O(1) (theta)
    - IDEA:
        - start small: m = 8 (some constant)
        - grows/shrinks as necessary
        - Let's say we insert 8 items and try inserting a 9th
        - If n > m: grow the table
    - Grow table: means that m --> m'
        - to do this we need to
            - make table of size m'
            - build new hash function h'
                - because the table has changed size
            - rehash:
                - for item in T:
                    - T'.insert(item) # in new table
            - this takes O(n + m + m') but is basically O(n)
        - m' = 2m --> **TABLE DOUBLING**
            - cost of n inserts:
                - O(1 + 2 + 4 + 8 + ... + n)
                - = O(n)
        - IDEA of AMORTIZATION:
            - operation takes "T(n) amortized" if k operations take <= k * T(n) time (at most k * T(n))
            - think of this as meaning "T(n) on average" where the average is over all operations
            - only relevant to data structures
        - IDEA: **Table Doubling**
            - for k inserts take O(k) time
                - O(1) amortized for insert
            - Also the same for k inserts and deletes they take O(k) time
        - Problem: if we do n inserts then n deletes but m = maximum value of n over time
            - Fix deletion:
                - 1. if m = n/2 then shrink to m/2
                    - ISSUE: this is slow if we go from 2^k <-> 2^k + 1 we get O(n) per operation
                - 2. if m = n/4 then shrink to m/2
                    - this works and the amortized time becomes O(1) constant

- String matching
    - given 2 strings s and t: does s occur as a substring of t?
    - simple algorithm --> O(|s| * |t|) in notes
- Rolling hash ADT:
    - given hash value r, we want to do r.append(c)
        - this adds character c to the end of x
        - r maintains a string x
    - we also want to delete the first character of x r.skip(first_char)
    - r() will give us the hash value of x = h(x)
- Karp-Rabin algorithm:
    - compute the hash function of search string s
        - note: rolling hash data structure only allows us to add characters
        - let our hash is rs
    - compute the hash function of the first s characters of our string t
        - let our hash be rt
    - check if the 2 hashes rs and rt are equal
        - just because the hashes are equal, doesn't mean the strings are equal.
        - we need to check whether or not s == the substring by iterating over s
        - if equal, we are done
        - else: we want to enginneer this so that this occurs at a low probability
    - iterate over len(s), len(t)
        - skip over the first letter
        - add one more additional letter
        - if rs and rt are equal
            - we need to check whether or not s == the substring by iterating over s
            - if equal, we are done
            - else: we want to enginneer this so that this occurs at a low probability
    - complexity: O(|s| + |t| + (# of matches) * |s|) which is linear O(1) if we care about a single match
