# 7. Counting Sort, Radix Sort, Lower Bounds for Sorting

- Linear time sorting:
    - counting sort
    - radix sort
- Model of Computation --> Comparison model:
    - all input items are black boxes (abstract data types)
    - only operations allowed are comparisons (<, <=, >, >=, ==)
    - time cost of algorithm in this model = # of comparisons
- Decision Tree:
    - any comparison algorithm can be viewed as a tree of all possible comparisons and their outcomes and their resulting answer
    - not practical but can be useful for algorithmic analysis
    - Analogues:
        - internal node in DT --> binary decisiion in algorithm (comparison)
        - leaf in DT --> found answer in algorithm
        - downward root to leaf path in DT --> algorithm execution
        - length of the path in DT --> running time of algorithm
        - height of tree in DT --> worst case running time
- Searching lower bound:
    - with the above models, we can prove that the worst case time to find a given item has a lower bound of O(log(n)), not necessarily tight
- **Linear-time sorting** --> INTEGER SORTING (assumption)
    - assumption is that the n keys we are sorting are integers
        - even if not sorting integers, can map things to integers
    - assume that integers are in {0, 1, ..., k-1} (k is max) and each integer fits in a word
    - can do a lot more than comparisons
    - for k not too big, we can sort in O(n) time
- **Counting Sort**
    - does not making comparisons
    - it counts all the items in an array 
    - Steps:
        - [Video explanation](https://www.youtube.com/watch?v=OKd534EWcdk)
        - start with array (new_array) of size k (# of distinct numbers in array)
        - traverse given array and write number of occurences at corresponding number index
        - in new_array, add each # to the right of it, accumulating the values
        - shift all elements in new_array to the right by 1 cell
        - now you can output the values
    - Complexity --> O(n + k)
        - if k = n, then O(n)
    - Note:
        - counting sort only sorts the integers, but items to be sort could have more data attached to them
- **Radix Sort**
    - [Video explanation](https://www.youtube.com/watch?v=XiuSW_mEn7g)
    - break each integer into a bunch of columns by imagining each integer as base b (not to be computed)
    - d = # of digits = log_b(k) + 1
    - sort integers by least significant digit ... sort integers by most significant digit
        - each of these sorts are done using _counting sort_ by digit
        - this step has O(n+b) complexity
    - Total time:
        - O((n+b) * d)
        - O((n+b) log_b(k)) but now we want to minimize this by setting b = n
        - O(n log_n(k)) and if k <= n^c, then the complexity is O(nc)
            - this means that if integers are reasonably small, we achieve a linear time sorting algorithm
    - Note:
        - larger base b --> more space required
