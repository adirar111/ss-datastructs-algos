# 20. Dynamic Programming II: Text Justification, Blackjack

- 5 easy steps (general):
    - 1. define subproblems --> # of subproblems
    - 2. guess (part of the solution) --> # of choices for guess
    - 3. relate subproblem solutions (with recurrence) --> # time per subproblem (similar to the # of choices for guess)
    - 4. Build algorithm doing either: --> ensure that the subproblem recurrence is acyclic i.e. in topological order (without needing to actually do topological sort).
        - recurse and memoize
        - build DP table (bottom-up approach)
    - 5. solve original problem --> consider extra running time
