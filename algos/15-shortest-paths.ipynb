# 15. Single-Source Shortest Paths Problem

- Will be exploring graphs with weights on their edges
    - Graphs are G(V, E, W)
        - V = vertices
        - E = edges
        - W = weights
- Algorithms:
    - Djikstra's Algorithm --> complexity is O(V * log(V) + E) essentially linear
        - only works on positive weight edges
        - E is at most V^2 or O(V^2)
    - Bellman-Ford Algorithm --> complexity is O(VE)
        - works on positive and negative weight edges
        - complexity is worse than Djikstra's
        - this needs to find cycles of negative weights which needs more work
    - Note that the complexities are independent of W (weight)
- Definitions:
    - path p = <v0, v1, ..., vk>
        - is path if (vi, vi+1) in E for 0 <= i < k 
    - Weight(path) = summation of weight of edges
    - Goal: find path p with minimum weight
    - Note: potentially exponential number of paths
        - algorithms will need to work around this cleverly
- Weighted graphs:
    - v0 -P-> vk
        - (v0) is a path from v0 to v0 of weight 0
    - shortest path weight from u to v as
        - S(u, v) where S is delta
            - either the minimum of weighted paths from u to v
            - or if no such path exists, denote the shortest path by infinity
- Things we need:
    - d(v): current weight of path at a vertex
        - whe d(v) becomes S, we found the best path
    - PI(v): predecesor vertex on the best path of v
        - PI(s) = None (the starting point)
- Negative weights:
    - why? examples:
        - paths that might lead to gain or loss in money
        - monopoly
        - social networks (likes, dislikes)
    - we can also think if there's a way to shift all weights such that we only deal with positive weights so that we can use Djikstra instead of BF
- Negative cycles:
    - may lead to infinite  --> PROBLEM
    - BF needs to give delta numbers for vertices that have finite numbers and then mark all other vertices with -inf
    - BF also needs to be able to detect negative cycles
- General structure of shortest path algorithms for no-negative cycles:
    - Initialize for v in V
        - d[v] = inf
        - PI[v] = None
        - d[s] = 0 --> source
    - Repeat
        - select edge (u, v) [somehow]
        - "Relax" edge (u, v) --> look at it, examine, and do things
            - if d[v] > d[u] + w(u, v)
                - do d[v] = d[u] + w(u, v)
                - do PI[v] = u
        - repeat until all edges have d[v] <= d[u] + w(u, v)
            - we want to reach the point where no edges can be relaxed
- Optimal substructure property:
    - Subpaths of a shortest path are shortest paths
    - see notes for triangle inequality
