# 16. Dijkstra

- Relax operation: Rela(u, v, w)
    - if d[v] > d[u] + w(u, v)
        - d[v] = d[u] + w(u, v)
        - PI(v) = u 
    - Relaxation is safe:
        - doing relaxation will give us the correct values from top to bottom
- Algorithm for shortest path on DAG (directed acyclic graph - NO CYCLES)
    - DAGs are allowed to have negative edges
    - Remember that when the search starts, all other vertices have distance of infinity but starting value has d[s] = 0
    - STEPS:
        - toplogical sort the DAG. Path from u to v implies that u is before v in the ordering
        - One pass (left to right) over vertices in topologically sorted order relaxing each edge that leaves each vertex
    - Complexity:
        - O(V+E) time
- Djikstra's algorithm:
    - greedy algorithm -- does things incrementally and iteratively builds shortest path
    - PSUEDOCODE:
        - ```
        Djikstra(G, W, s): # G is graph, W are weights, s is starting vertex
            Initialize (G, s) --> sets d[s] = 0
            Set S = None --> contains vertices we know the shortest path to
            Priority queue Q = V[G] --> gets the entire set of vertices from the graph
                                    --> priority is based on d values of vertices
            while Q != None:
                u = extract_min(Q) --> deletes u from Q too
                S = S union {u} --> adds u to S
                for each vertex in Adj[u]: --> for all vertices that point out of u
                    Relax(u, v, w) --> relax subroutine
            ```
    - once an element is in S, it's not going to reduce anymore
    - Complexity:
        - Priority queue costs:
            - O(V) inserts into the priority queue
            - O(V) extract min operations
            - O(E) decrease-key operations (relax operation decreases key, reducing d value)
                - only processing each edge out of a vertex once
        - If using array as priority queue:
            - O(V) = cost of extract min
            - O(1) = cost of decrease key
            - Total: O(V * V + E * 1) = O(V^2) because E is order V^2
        - If using min heap:
            - O(log(V)) = cost of extract min
            - O(log(V)) = cost for decrease key
            - Total: O(Vlog(V) + Elog(V))
        - To obtain optimal djikstra, implement priority queue using fibonacci heap
            - O(log(V)) = cost of extract min
            - O(1) amortized time for decrease key
            - Total: O(Vlog(V) + E)
