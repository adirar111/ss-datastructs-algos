# 19. Dynamic Programming I: Fibonacci, Shortest Paths

- Motivation: desigining new algorithms
- Dynamic programming (DP) is a general and powerful algorithm design technique
    - "careful brute force"
- Examples explained using fibonacci and shortest paths problem
- DP - Memoization:
    - memoize (remember):
        - remember and reuse solutions to subproblems that help solve the problem.
    - time = # subproblems * (time per subproblem)
    - subproblem dependencies should be acyclic
        - to make cyclic acyclic:
            - make multiple layers/ copies of the graph, reduce it to k copies.
            - every edge takes you down to the next layer
            - <img src="screenshots/img_6.png" width="300" />
- DP - Bottom-up DP:
    - does exactly the same computation as computation
    - topological sort of subproblem dependency DAG
        - topological sort is usually just a for loop
        - need to move in order to compute what you want
    - can often save space
